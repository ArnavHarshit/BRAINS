```python
# import the necessary packages
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
import numpy as np

plt.style.use('seaborn-v0_8-darkgrid')
pd.set_option('display.max_columns', None)
```

```python
# set the start and end dates for our market data request
end_date = datetime(year=2025, month=3, day=1)
start_date = datetime(year=2020, month=1, day=1)
```

```python
# set the name of the ticker we want to download market data for
ticker = ["BTC-USD"]

# download market data for a single ticker
df = yf.download(
    tickers=ticker,
    start=start_date,
    end=end_date,
    interval="1d",
    group_by="ticker",
    auto_adjust=True,
    progress=False
)
df['BTC-USD','Close']
df.shape[0]
```

```python
# initialize a new figure
plt.figure(figsize=(14, 7))
# sns.set(style="whitegrid")

# loop over the tickers
# for ticker in tickers:
    # plot the closing price for each
sns.lineplot(
        data=df['BTC-USD','Close'],
        label=ticker,
        linewidth=2
    )

# set the plot title
plt.title(
    f"Stock Closing Prices ("
    f"{start_date.strftime('%Y-%m-%d')} "
    f"to {end_date.strftime('%Y-%m-%d')})"
)

# set the plot labels
plt.xlabel("Date")
plt.ylabel("Closing Price ($)")

# finish constructing the plot
plt.tight_layout()
plt.show()
```

```python
df.reset_index(inplace=True)
```

```python
fig = go.Figure(data=[
    go.Candlestick(x=df.index,
                   open=df['BTC-USD','Open'],
                   high=df['BTC-USD','High'],
                   low=df['BTC-USD','Low'],
                   close=df['BTC-USD','Close'])
              ])
fig.update_layout(title=f"Candlestick Chart - {ticker}", xaxis_title='Date', yaxis_title='Price')
fig.show()
```

```python
df['Daily Return'] = df['BTC-USD','Close'].pct_change() # Returns 0.01 for 1% change
arr = []
arr2 = []
positivereturns = 0
negativereturns = 0
for i in range(1,1885):
    if((df['BTC-USD','Close'][i] - df['BTC-USD','Close'][i-1])>0):
        positivereturns += (df['BTC-USD','Close'][i] - df['BTC-USD','Close'][i-1])
    else :
        negativereturns += -(df['BTC-USD','Close'][i] - df['BTC-USD','Close'][i-1])
    change = (df['BTC-USD','Close'][i] - df['BTC-USD','Close'][i-1])/df['BTC-USD','Close'][i-1]
    log_change = np.log(df['BTC-USD','Close'][i]/df['BTC-USD','Close'][i-1])
    arr.append(change) # manually calculating the change
    arr2.append(log_change)
```

```python
# Visualize daily returns
plt.figure(figsize=(12, 6))
plt.bar(df.index, df['Daily Return']*100, color='green')
plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
plt.title(f'{ticker} Daily Returns (%)')
plt.xlabel('Date')
plt.ylabel('Return (%)')
plt.grid(True)
plt.show()
```

```python
df['Log Return'] = np.log(df['BTC-USD','Close'] / df['BTC-USD','Close'].shift(1))
df[['Daily Return', 'Log Return']].dropna().head()
```

```python
fig1 = px.line(df, x=df.index, y=df['Daily Return'], title=f'{ticker} Simple Returns')
fig1.show()

fig2 = px.line(df, x=df.index, y=df['Log Return'], title=f'{ticker} Log Returns')
fig2.show()
```

```python
df['Rolling Std Dev'] = df['Daily Return'].rolling(window=21).std()
fig = px.line(df, x=df.index, y=df['Rolling Std Dev'], title='Rolling 21-Day Standard Deviation')
fig.show()
```
