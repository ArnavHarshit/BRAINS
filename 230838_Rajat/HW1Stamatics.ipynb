```python
# import the necessary packages
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
# set the start and end dates for our market data request
end_date = datetime(year=2025, month=3, day=1)
start_date = datetime(year=2024, month=3, day=1)
```

```python
# set the name of the ticker we want to download market data for
ticker = "AAPL"

# download market data for a single ticker
df_single = yf.download(
    tickers=ticker,
    start=start_date,
    end=end_date,
    interval="1d",
    group_by="ticker",
    auto_adjust=True,
    progress=False
)
df_single
```

```python
df_single['Simple return']=df_single['AAPL']['Close'].pct_change()
df_single['Simple return']
```

```python
import numpy as np
df_single['Log returns']=np.log(df_single['AAPL']['Close']/df_single['AAPL']['Close'].shift(1))
df_single['Log returns']
```

```python
# Calculate the rolling 14-day standard deviation of the log returns
df_single['Rolling_Volatility'] = df_single['Log returns'].rolling(window=14).std()

# Create a figure and twin axes
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()

# Plot the closing price on the primary axis
ax1.plot(df_single.index, df_single['AAPL']['Close'], label='Close Price', color='blue')
ax1.set_ylabel('Close Price')

# Plot the rolling volatility on the secondary axis
ax2.plot(df_single.index, df_single['Rolling_Volatility'], label='Rolling 14-Day Volatility', color='red')
ax2.set_ylabel('Rolling Volatility')

# Add a title and legends
plt.title('AAPL Close Price and Rolling 14-Day Volatility')
fig.legend(loc="upper left")

# Show the plot
plt.show()
```

```python
# Filter the DataFrame to include only rows where 'Class' is 'UP'
df_up_days = df_single[df_single['Class'] == 'UP']

# Calculate the average of the 'Simple return' column for these rows
average_simple_return_up = df_up_days['Simple return'].mean()

# Print the result
print(f"The average simple return on days when the Class is 'UP' is: {average_simple_return_up}")
# Filter the DataFrame to include only rows where 'Class' is 'DOWN'
df_down_days = df_single[df_single['Class'] == 'DOWN']

# Calculate the average of the 'Simple return' column for these rows
average_simple_return_down = df_down_days['Simple return'].mean()

# Print the result
print(f"The average simple return on days when the Class is 'DOWN' is: {average_simple_return_down}")
```
